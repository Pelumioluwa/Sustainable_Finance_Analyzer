## This splits the csv into training and validation data
This includes the following steps
1. import the csv
2. separate the sustainable finance and non-sustainaable finance text in different dfs.
3. shuffle both dfs and take the first 5000 of both dfs as training data set. Save 10,000 texts into a csv
4. save the remaining as validation data sets


```python
#import needed libraries
import pandas as pd
```

```python
csv_directory = '/Users/pelumioluwaabiola/Downloads/Researchwork/CSVfiles/standards.csv'
test_train_directory = '/Users/pelumioluwaabiola/Downloads/Researchwork/CSVfiles'
```

```python
df = pd.read_csv(csv_directory)
df.head()
```

```python
#get total number of rows in df
total_rows = df.shape[0]
print('Total number of rows in df: ', total_rows)
```

```python
#retrieve values with label = 0 and save in new df
sustainable_df = df[df['label']==0]
rows = sustainable_df.shape[0]
print('Number of sustainable standards:', rows)
```

```python
#retrieve values with label = 1 and save in a new df
nonsustainable_df = df[df['label']==1]
num_rows = nonsustainable_df.shape[0]
print('Number of non-sustainable standards:', num_rows)
```

```python
#view the head rows before the shuffle
sustainable_df.head()
```

```python
#shuffle and view head rows after the shuffle
sustainable_df = sustainable_df.sample(frac=1).reset_index(drop=True)
sustainable_df.head()
```

```python
#view head of nonsustainable_df before shuffle
nonsustainable_df.head()
```

```python
#shuffle and view head of nonsustainable_df after shuffle
nonsustainable_df = nonsustainable_df.sample(frac=1).reset_index(drop=True)
nonsustainable_df.head()
```

```python
#get the first 5000 rows and save in a new df
training_sustainable_df = sustainable_df.iloc[:7000]
print('Number of rows in training_sustainable_df:', training_sustainable_df.shape[0])
```

```python
training_sustainable_df.head()
```

```python
training_sustainable_df.tail()
```

```python
#get the first 5000 rows and save in a new df
training_nonsustainable_df = nonsustainable_df.iloc[:7000]
print('Number of rows in training_nonsustainable_df:', training_nonsustainable_df.shape[0])
```

```python
training_nonsustainable_df.head()   
```

```python
#merge both traiing_sustainable_df and training_nonsustainable_df
training_df = pd.concat([training_sustainable_df, training_nonsustainable_df])
print('Number of rows in training_df:', training_df.shape[0])
```

```python
#shuffle the training_df
training_df = training_df.sample(frac=1).reset_index(drop=True)
training_df.head()
```

```python
#get the get remaining rows after the 7000 and save in a new df
test_sustainable_df = sustainable_df.iloc[7000:]
print('Number of rows in test_sustainable_df:', test_sustainable_df.shape[0])
```

```python
test_sustainable_df.head()
```

```python
#get the get remaining rows after the 5000 and save in a new df
test_nonsustainable_df = nonsustainable_df.iloc[7000:]
print('Number of rows in test_nonsustainable_df:', test_nonsustainable_df.shape[0])
```

```python
test_sustainable_df.head()
```

```python
#merge both test 
test_df = pd.concat([test_sustainable_df, test_nonsustainable_df])
print('Number of rows in test_df:', test_df.shape[0])
```

```python
#shuffle the test_df
test_df = test_df.sample(frac=1).reset_index(drop=True)
test_df.head()
```

```python
#write training data in csv file
document_title = '/training.csv'
training_df.to_csv(test_train_directory + document_title, index=False)
```

```python
#write test data in csv file
document_title = '/test.csv'
test_df.to_csv(test_train_directory + document_title, index=False)
```
